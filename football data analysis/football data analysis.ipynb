```python
# Import the liblaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
epl_df = pd.read_csv("England Premier League Players 2022-2023.csv")
```

```python
# Create 2 new columns
#epl_df['MinsPerMatch'] = (epl_df['Min']/ epl_df['90s']).astype(float)
#epl_df['GoalsPerMatch'] =(epl_df['Gls']/ epl_df['MP']).astype(float)

epl_df["Mins Per Match"] = ((epl_df["Min"]/ epl_df["90s"]).astype(float)).round(2)
epl_df["Goals Per Match"] =((epl_df["Gls"]/ epl_df["MP"]).astype(float)).round(2)

epl_df.head()
```

```python
# Total Goals
total_goals = epl_df['Gls'].sum()
print(total_goals)
```

```python
#PK -- Penalty Kicks Made
#Penalty Kicks Made
total_penalty_kicks_made = epl_df['PK'].sum()
print(total_penalty_kicks_made)
```

```python
#PKatt -- Penalty Kicks Attempted
#Penalty Kicks Attempted
total_penalty_kicksa_attempted = epl_df['PKatt'].sum()
print(total_penalty_kicksa_attempted)
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
pl_not_scored = epl_df['PKatt'].sum() - total_penalty_kicks_made
data = [pl_not_scored, total_penalty_kicks_made]
labels = ['Penalties missed', 'Penalties scored']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Total assists from each club
assists_by_clubs = pd.DataFrame(epl_df.groupby("Squad", as_index=False)["Ast"].sum())
sns.set_theme(style="whitegrid", color_codes = True)
ax = sns.barplot(x="Squad", y="Ast", data = assists_by_clubs.sort_values(by="Ast"), palette = "Set2")
ax.set_xlabel("Squad", fontsize=30)
ax.set_ylabel("Assists", fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title("Plot of Clubs vs Total Assists", fontsize=20)
```

```python
# Top 10 Assists
top_10_assists = epl_df[["Player", "Squad", "Ast", "MP"]].nlargest(n=10, columns = "Ast")
top_10_assists.reset_index()
```

```python
# Total goals from each club
goals_by_clubs = pd.DataFrame(epl_df.groupby("Squad", as_index=False)["Gls"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Squad", y="Gls", data = goals_by_clubs.sort_values(by="Gls"), palette='rocket')
ax.set_xlabel("Squad", fontsize=30)
ax.set_ylabel("Goals", fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title("Plot of Clubs vs Total Goals", fontsize=20)
```

```python
# Most goals by players
top_10_goals = epl_df[["Player", "Squad", "Gls", "MP"]].nlargest(n=10, columns = "Gls")
top_10_goals.reset_index()
```
